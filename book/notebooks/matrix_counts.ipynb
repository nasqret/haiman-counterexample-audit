# Reproduce the 90-row and 115-column counts

This notebook translates the six printed row families and two printed column families literally.

```python
from itertools import combinations

rows = []
row_family_sizes = []
for a in range(1, 4):
    families = [
        [(a, j, i, k) for j, i, k in combinations(range(4, 9), 3)],
        [(a, j, i, k) for k, j, i in combinations(range(4, 9), 3)],
        [(a, 4, 4, k) for k in range(5, 9)],
        [(a, j, 4, j) for j in range(5, 9)],
        [(a, 5, 5, 6)],
        [(a, 6, 5, 6)],
    ]
    if not row_family_sizes:
        row_family_sizes = [len(family) for family in families]
    for family in families:
        rows.extend(family)

columns_1 = [(r, s, t) for r in range(1, 4) for s in range(1, 4) for t in range(4, 9)]
columns_2 = [
    (r, s, t)
    for r in range(4, 9)
    for s in range(4, 9)
    for t in range(s, 9)
    if not (r == s == t)
]
columns = columns_1 + columns_2

assert row_family_sizes == [10, 10, 4, 4, 1, 1]
assert len(rows) == len(set(rows)) == 90
assert len(columns_1) == 45
assert len(columns_2) == 70
assert len(columns) == len(set(columns)) == 115

{
    'row_family_sizes_per_a': row_family_sizes,
    'rows': len(rows),
    'columns_block_1': len(columns_1),
    'columns_block_2': len(columns_2),
    'columns': len(columns),
}
```
